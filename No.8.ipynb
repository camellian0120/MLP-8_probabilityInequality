***正規分布に近似した分布の作成***

```python
# 正規分布の作成
# 参考 : https://qiita.com/Qwertyutr/items/59e189d287ba6e689ed9
# imports
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# おおよそ正規分布となる値を生成する関数
def mkNearNormalDistribution(size=500):
    # numpyで基準の乱数生成?
    randomBase = np.random.default_rng()

    # numpyで先ほど決めた値に対して、正規分布として従う乱数の生成
    RandomValue = randomBase.normal(loc=5, scale=1, size=size)

    # 生成した値を返す
    return RandomValue

# ここでのみテスト実行
if "__main__" in __name__:
    test = mkNearNormalDistribution()

    # seabornで密度分布をプロット
    sns.displot(test, kind='kde')

    # 配列の中身の確認
    # print(test)
```

***マルコフ不等式とチェルノフ不等式の実装\
\( ここで、ほぼ正規分布であることから上記のloc=5より、期待値E\[X\]=5とする。\)***

```python
# マルコフ不等式・チェルノフ不等式
# need import seaborn as sns
from math import exp

# マルコフ不等式が成立するかを確かめる関数
def Markov(graph, size=-1, alpha=-1,Ex=0):
    if alpha >= 0 and size > 0:
        # alphaより大きな値を記録する変数
        count = 0
        # alphaより大きな値を探す
        for loop in graph:
            if loop >= alpha:
                count += 1
        # 結果の表示
        print("P(x >= "+str(alpha)+") = "+str(count/size))
        print("E[x]/a = "+str(5/alpha))

    else:
        print("3つの引数は必ず指定してください。\ngraphは配列要素、sizeは配列の大きさ、alphaは調べたい上側確率です。")

# チェルノフ不等式が成立するかを確かめる関数
def Chernoff(graph, size=-1, alpha=0, Ex=0, t=1):
    if size > 0:
        # 絶対値sigmaより遠くにいる値を記録する変数, また積率母関数Mx(t)の計算用の Mxt
        count = 0
        Mxt = 0
        # sigmaより離れている値を探す
        for loop in graph:
            Mxt += exp(loop * t)/500
            if exp(loop * t) >= exp(alpha * t):
                count += 1
        # 結果の表示
        print("P(x >= "+str(alpha)+") = "+str(count/size))
        print("E[x]/a = "+str(Mxt/exp(alpha * t)))

    else:
        print("3つの引数は必ず指定してください。\ngraphは配列要素、sizeは配列の大きさ、sigmaは調べたい上側確率及び下側確率です。")

# 今回だけ実行
if "__main__" in __name__:
    # 関数でほぼ正規分布の作成
    size = 500
    test = mkNearNormalDistribution(size)

    # マルコフ不等式の検証
    print("## マルコフ不等式が正しいかを確認する(E[x] = 5 の正規分布に近い連続確率分布)")
    for loop in range(5, 8):
        print("# case a="+str(loop))
        Markov(graph=test, size=size, alpha=loop, Ex=5)

    # チェルノフ不等式の検証
    print("## チェルノフ不等式が正しいかを確認する(E[x] = 5 の正規分布に近い連続確率分布)")
    for loop in range(1, 11):
        print("# case t="+str(loop))
        Chernoff(graph=test, size=size, alpha=7, Ex=5, t=loop)

    # seabornで密度分布をプロット
    sns.set(style="ticks")
    sns.displot(test, kind='kde')
```
