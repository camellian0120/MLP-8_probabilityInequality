## Jensen's Inequality program

```python
# imports
import numpy as np
from scipy.stats import norm #normは正規分布
import matplotlib.pyplot as plt

## variables
# X scale
X = np.arange(-3, 3, 0.1)
```

### graph plot

```python
# 参考 : https://lib-arts.hatenablog.com/entry/VariationalInference2
def Jensen(fx=X**2, atob=30):
    # 範囲[a, b]
    # a~b間の距離 length
    length = atob
    a = [X[10],        fx[10]]
    b = [X[10+length], fx[10+length]]

    # 変化の割合 rate
    rate = (b[1]-a[1])/(length)
    chord_x = np.arange(a[0], b[0], 0.1)
    chord_y = np.arange(a[0], b[0], 0.1)
    for loop in range(len(chord_x)):
        chord_y[loop] = a[1]+loop*rate

    # plot
    # lines
    plt.plot(X,       fx,      color='orange')
    plt.plot(chord_x, chord_y, color='cyan')
    # dots
    plt.plot(a[0], a[1], marker='o', color='cyan')
    plt.plot(b[0], b[1], marker='o', color='cyan')

if "__main__" in __name__:
    Jensen()
```

```python
Jensen(fx=X**3+3*X**2+3**X+1, atob=40)
```

```python
Jensen(fx=2*np.sin(X), atob=40)
```
