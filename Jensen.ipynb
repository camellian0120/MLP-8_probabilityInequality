## Jensen's Inequality program

```python
# imports
import numpy as np
from scipy.stats import norm #normは正規分布
import matplotlib.pyplot as plt

## variables
# X scale
X = np.arange(-3, 3, 0.1)
```

### graph plot

```python
# 参考 : https://lib-arts.hatenablog.com/entry/VariationalInference2
# set value
# Y scale
fx = X**2

# 範囲[a, b]
a = [X[10], fx[40]]
b = [X[10], fx[40]]

chord_x = np.arange(a[0], b[0]+0.1, 0.1)
chord_y = chord_x
for loop in range(len(chord_y)):
    chord_y[loop] = (chord_x)

# plot
plt.plot(X, fx, color='r')
plt.plot(chord_x, chord_y)
```
